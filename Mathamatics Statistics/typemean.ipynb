```python
symbols = ['BTC-USD','BND']
start_date = '2023-01-01'
end_date = '2023-12-31'

df = pd.DataFrame()

for symbol in symbols:
    df[symbol] = yf.download(symbol,start=start_date,end=end_date)['Close']

returns = df.pct_change().dropna()
```

```python
def rms(series):
    return np.sqrt(np.mean(np.square(series)))

btc_weights = np.arange(0.1,1.01,0.1)
rms_values = []

for w in btc_weights:
    weights = [w,1-w]

    portfolio_returns = returns.dot(weights)
    rms_val = rms(portfolio_returns) * 100
    rms_values.append(rms_val)
```

```python
plt.figure(figsize = (8,5))
plt.plot(btc_weights * 100, rms_values, marker='o', color='darkorange')
plt.title('Portfolio RMS Volatilty vs BTC Allocation (2023)')
plt.xlabel('BTC Allocation (%)')
plt.ylabel('RMS Volatility (%)')
plt.grid(True,linestyle='--',alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
def rolling_rms(series,window):
    return series.rolling(window).apply(lambda x: np.sqrt(np.mean(np.square(x))), raw=True)

rolling_window = 30
rms_rolling = pd.DataFrame()
for symbol in symbols:
    rms_rolling[symbol] = rolling_rms(returns[symbol],rolling_window) * 100
```

```python
plt.figure(figsize=(10, 5))
for symbol in symbols:
    plt.plot(rms_rolling.index, rms_rolling[symbol], label=symbol)

plt.title('30-Day Rolling RMS Volatility (2023)')
plt.ylabel('RMS Volatility (%)')
plt.xlabel('Date')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
def simulate_portfolio(weights_dict,start='2023-01-01',end='2023-12-31',plot=True):
    symbols = list(weights_dict.keys())
    weights = list(weights_dict.values())

    df = pd.DataFrame()
    for symbol in symbols:
        df[symbol] = yf.download(symbol, start=start,end=end)['Close']

    returns = df.pct_change().dropna()

    portfolio_returns = returns.dot(weights)

    def rms(series):
        return np.sqrt(np.mean(np.square(series)))
    
    rms_values = {symbol: rms(returns[symbol]) * 100 for symbol in symbols}
    rms_values['Portfolio'] = rms(portfolio_returns) * 100

    if plot:
        plt.figure(figsize=(9, 5))
        bars = plt.bar(rms_values.keys(), rms_values.values(), color=['gold', 'steelblue', 'mediumseagreen'])
        plt.title(f"RMS Volatility: Portfolio vs Assets ({start[:4]})")
        plt.ylabel('RMS Volatility (%)')
        for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, yval + 0.2, f'{yval:.2f}%', ha='center')
        plt.grid(axis='y', linestyle='--', alpha=0.6)
        plt.tight_layout()
        plt.show()
    
    return rms_values

```

```python
simulate_portfolio({'AAPL':0.2,'TSLA':0.1,'BTC-USD':0.3,'BND':0.4})
```

```python
symbols = ['SPY','RSP']
start_date = '2023-01-01'
end_date = '2023-12-31'

df = pd.DataFrame()

for symbol in symbols:
    df[symbol]= yf.download(symbol,start=start_date,end=end_date)['Close']

returns = df.pct_change().dropna()
```

```python
cumulative_returns = (1+returns).cumprod()

plt.figure(figsize=(10,5))
for symbol in symbols:
    plt.plot(cumulative_returns.index, cumulative_returns[symbol], label=symbol)

plt.title('SPY vs RSP: Cumulative Returns in 2024')
plt.xlabel('Date')
plt.ylabel('Growth of $1 investment')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
score = [80,90,75]
credits = [4,3,2]

weighted_mean = np.average(score,weights=credits)
weighted_mean_2 = sum(s * c for s,c in zip(score,credits)) / sum(credits)
print(weighted_mean)
print(weighted_mean_2)
```
